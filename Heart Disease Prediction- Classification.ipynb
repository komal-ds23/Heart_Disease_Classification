## Domain Analysis :

1. __slope_of_peak_exercise_st_segment :__ The slope of the peak exercise ST segment, an electrocardiography read out indicating quality of blood flow to the heart. Flat, down sloping, or depressed ST segments may indicate "coronary ischemia", also called "Myocardial ischemia".1 : upsloping, 2: flat, 3: down sloping.•	For instance, ST elevation with an upsloping ST segment is generally considered normal, while ST elevation with a horizontal ST segment is more characteristic of myocardial ischemia.The current Third Universal Definition of Myocardial Infarction document states that horizontal or downsloping ST depression is a sign of ischemia, whereas upsloping ST depression pattern is not specified as indicative of ischemia. A downsloping ST depression suggests the most extensive ischemia, while horizontal remains as the marker for a moderate process and an upsloping pattern indicates significant but smaller size.

2. __thal :__ Thalassemia is an inherited blood disorder caused when the body doesn't make enough of a protein called hemoglobin, an important part of red blood cells. Results of thallium stress test measuring blood flow to the heart, with possible values normal, fixed_defect, reversible_defect. A normal result means blood flow through the coronary arteries is probably normal. 1: fixed defect (no blood flow in some part of the heart), 2: normal blood flow, 3: reversible defect (a blood flow is observed but it is not normal)

3. __resting_blood_pressure :__ Usually, blood pressure starts to rise a few hours before a person wakes up. It continues to rise during the day, peaking in midday. Blood pressure typically drops in the late afternoon and evening. Blood pressure is usually lower at night while sleeping. Normal range for blood pressure is less than 120/80 millimeters of mercury (mmHg).

4. __chest_pain_type :__ Angina is the term for chest pain caused by poor blood flow to the heart. Which is chest pain caused by blockages in the blood vessels leading to your heart. __Typical angina__ - classic chest pain that is often described as pressure or squeezing, __Atypical angina__ - chest pain with different or less typical symptoms, __Non-anginal chest pain__ - chest pain not caused by angina, __Asymptomatic__ - no chest pain or other symptoms present. 0: typical angina, 1: atypical angina, 2: non-anginal pain, 3: asymptomatic.

5. __num_major_vessels :__ In the context of coronary angiography, the number of major vessels (0-3) refers to the number of coronary arteries with significant stenosis (narrowing) that are visualized during the exam. A value of 0 indicates that no major vessels are visualized, while a value of 3 indicates that all three major coronary arteries are visualized. The values in between represent the number of major vessels that are visualized. It's important to note that __"fluoroscopy"__ is only one of the methods used to assess the number of major vessels, and the actual number may differ based on the method used and the specific patient and disease condition.

6. __fasting_blood_sugar_gt_120_mg_per_dl :__ Fasting blood sugar greater than 120 mg/dl (yes / no). Normal fasting blood sugar levels for adults without diabetes are typically in the range of 70-99 mg/dL. However, individual levels may vary and a healthcare provider can provide specific guidance based on personal health and medical history. If blood sugar levels are consistently higher than 100 mg/dL while fasting, it may indicate prediabetes or diabetes and further evaluation by a healthcare provider is recommended.

7. __resting_ekg_results :__ Resting electrocardiographic results (values 0,1,2). Value 0: normal, Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV), Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria. Nonspecific ST-segment and T-wave (ST-T) changes represent one of the most prevalent electrocardiographic abnormalities in hypertensive patients. Hypertrophy is the enlargement of an organ or tissue from the increase in size of its cells.

8. __serum_cholesterol_mg_per_dl :__ Serum cholesterol is a type of fat (lipid) found in the blood. It is an essential component of cell membranes and is necessary for the production of hormones, bile acids, and vitamin D. However, high levels of cholesterol in the blood can increase the risk of heart disease and stroke. There are two main types of cholesterol: low-density lipoprotein (LDL) cholesterol and high-density lipoprotein (HDL) cholesterol. LDL cholesterol is often referred to as "bad" cholesterol because it can build up in the walls of the arteries, forming plaques that can narrow or block the arteries. This increases the risk of heart disease and stroke. On the other hand, HDL cholesterol is often referred to as "good" cholesterol because it helps remove excess cholesterol from the bloodstream and can help prevent the buildup of plaques in the arteries. The total cholesterol level in the blood is the sum of LDL cholesterol, HDL cholesterol, and other lipids. A healthy total cholesterol level is generally considered to be below 200 milligrams per deciliter (mg/dL), while a level above 240 mg/dL is considered high. To assess the risk of heart disease, healthcare providers typically consider not only the total cholesterol level, but also the levels of LDL cholesterol and HDL cholesterol and other factors, such as the patient's age, gender, family history, and lifestyle factors.

9. __oldpeak_eq_st_depression :__ Oldpeak = ST depression induced by exercise relative to rest, a measure of abnormality in electrocardiograms. In the context of ECG and oldpeak/ST depression, a value greater than 4 millimeters (mm) can indicate significant myocardial ischemia, which is a reduced blood flow to the heart muscle. This can be a sign of underlying heart disease, such as coronary artery disease, and is generally considered to be a significant finding.

10. __max_heart_rate_achieved :__ Maximum heart rate achieved (beats per minute). Maximum heart rate (MHR) is the highest number of beats per minute (bpm) that your heart can safely handle during physical activity. It is used to determine exercise intensity and help monitor progress.<br>
To interpret MHR, you can use the following formula:<br>
MHR = 220 – Age<br>
The resulting number is an estimate of your maximum heart rate, and can be used to determine your target heart rate zone for exercise. The target heart rate zone is usually between 50-85% of your MHR, which can help you get an optimal workout without overexerting yourself. A heart rate that exceeds your maximum heart rate (MHR) by a significant amount, or a heart rate that remains elevated even after physical activity has stopped, can indicate an underlying issue such as cardiovascular disease, heart attack, or an electrolyte imbalance. It is also important to note that heart rates that are consistently and significantly above your MHR can lead to overexertion and increased risk of injury.

11. __exercise_induced_angina :__ •	exercise-induced chest pain (0: False, 1: True). Exercise-induced chest pain can be a sign of a potential heart problem or another underlying health issue. <br>
Some common causes of exercise-induced chest pain include:
* __Angina :__ Chest pain caused by reduced blood flow to the heart muscle.
* __Coronary artery disease :__ Narrowing or blockage of the blood vessels that supply blood to the heart muscle.
* __Aortic dissection :__ A tear in the inner wall of the aorta, the main blood vessel that carries blood away from the heart.
* __Pulmonary embolism :__ A blockage in the lung's blood vessels, often due to a blood clot.
* __Costochondritis :__ Inflammation of the cartilage that connects a rib to the breastbone.

## __Some Important User Defined Functions:__

__1. Cross val score__

```python
def UserCrossValScores(estimator, X, y, cv = None, n_jobs = -1, scoring = "accuracy"):
    if cv == None:
        cv = 5
    scores = cross_val_score(estimator = estimator, X = X, y = y, cv = cv, n_jobs = n_jobs, scoring = scoring)
    print(f"Scores : {scores}")
    print(f"Mean score : {np.mean(scores)}")
    print(f"Standard Dev. : {np.std(scores)}")
```

__2. Learning Curve__

```python
def UserLearningCurve(estimator, X, y, label,train_size = np.linspace(0.1, 1, 5), cv = None, n_jobs = -1, scoring = "accuracy"):
    if cv == None:
        cv = 5
    train_sizes, train_scores, validation_scores = learning_curve(estimator = estimator, 
                                                                  X = X, y = y, train_sizes = train_size, cv = cv, scoring = scoring, n_jobs = n_jobs)
    mean_train_score = np.mean(train_scores, axis = 1)
    mean_validation_score = np.mean(validation_scores, axis = 1)
    
    plt.plot(train_sizes, mean_train_score, label = "Training", marker = "o")
    plt.plot(train_sizes, mean_validation_score, label = "Validation", color = "g", marker = "o")
    plt.title(f"Learning curve for {label}")
    plt.xlabel("No. of train sizes")
    plt.ylabel(f"{scoring.capitalize()}")
    plt.legend(loc = "best")
```

__3. ROC_AUC__

```python
def UserRocAuc(y_true, y_prob, label = ""):
    fpr, tpr, thresholds = roc_curve(y_true, y_prob[:, 1])
    #auc_score = auc(fpr, tpr)
    auc_score = roc_auc_score(y_true, y_prob[:, 1])
    
    plt.plot(fpr, tpr, label = f"{label}(AUC = {round(auc_score, 5)})")
    
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve")
    
    plt.legend(loc = "best")
```

__4. Grid Search CV__

```python
def UserGridSearchCV(model, params, X, y, cv = None, n_jobs = -1, scoring = "accuracy"):
    if cv == None:
        cv = 5
    grid_cv = GridSearchCV(estimator = model, param_grid = params, cv = cv, scoring = scoring, n_jobs = n_jobs)
    
    grid_cv.fit(X, y)
    
    print(f"Best Params : {grid_cv.best_params_}")
    print(f"Best Score : {grid_cv.best_score_}")
    print(f"Best Estimator : {grid_cv.best_estimator_}")
    
    return grid_cv.best_estimator_
```

__5. Metric Evaluation__

```python
def UserMetric(y_true, y_pred, label):
    print(f"{label}:")
    print(f"Accuracy : {accuracy_score(y_true, y_pred)}")
    print(f"Precision : {precision_score(y_true, y_pred)}")
    print(f"Recall : {recall_score(y_true, y_pred)}")
    print(f"{classification_report(y_true, y_pred)}")
    print(f"Confusion Matrix : {confusion_matrix(y_true, y_pred).ravel()}")
    print("--"*40)
```

### Defining a function which will take user data and predict the Probabilty and predicted output if a person is having Heart Disease or not.

```python
def predict_heart_disease():
    """
    This function return the predicted chances of getting heart disease.
    
    Returns:
        int or float: The Probabilty along with predicted output.
    """
    
    sex = int(input("Enter your gender(0:female, 1:male):"))
    age = int(input("Enter your age:"))
    slope_of_peak = int(input("Enter the type of slope of peak ST segment(1 : upsloping, 2: flat, 3: down sloping):"))
    thal = int(input("Enter the result of thallium stress test measuring blood flow to the heart(1: fixed defect, 2: normal blood flow, 3: reversible defect):"))
    resting_bp = int(input("Enter the blood pressure at rest(mmHg):"))
    cp_type = int(input("Enter the chest pain type(1:typical angina,2:atypical angina,3:non-anginal pain,4:asymptomatic):"))
    major_vessels = int(input("Enter the number major vessels(0 - 3):"))
    fasting_blood_sugar = int(input("Fasting Blood Pressure greater than 120(mg/dl) [Yes(1)/No(0)]:"))
    resting_ekg = int(input("Resting electrocardiographic results(0: normal,1: having ST-T wave abnormality,2: showing probable or definite left ventricular hypertrophy):"))
    serum_cholesterol = float(input("Enter the amount of serum cholestrol(mg/dl):"))
    oldpeak_st_dep = float(input("Enter oldpeak/ST depression(mm):"))
    max_heart_rate = float(input("Enter the maximum heart rate:"))
    ex_induced_angina = int(input("Any chest pain due to exercise[Yes(1)/No(0)]:"))
    
    thal_value = ""
    if thal == 1:
        thal_value = "fixed_defect"
    elif thal == 2:
        thal_value = "normal"
    elif thal == 3:
        thal_value = "reversible_defect"
        
    test_data = [
        slope_of_peak,
        thal_value,
        resting_bp,
        cp_type,
        major_vessels,
        fasting_blood_sugar,
        resting_ekg,
        serum_cholesterol,
        oldpeak_st_dep,
        sex,
        age,
        max_heart_rate,
        ex_induced_angina
    ]
    
    #--------------Encoding, Scaling and Predicting the test data----------------------
    
    preprocessed_value = preprocessing_steps.transform([test_data])
    prediction = XGB_model.predict(preprocessed_value)[0]
    
    if prediction == 0:
        prediction_proba = XGB_model.predict_proba(preprocessed_value)[:, 0]
        print(f"\nNo Heart Disease and Chances are {round((prediction_proba[0]*100), 3)}%")
    elif prediction == 1:
        prediction_proba = XGB_model.predict_proba(preprocessed_value)[:, 1]
        print("\n*************************WARNING**********************************")
        print(f"You have a Heart Disease and Chances are {round((prediction_proba[0]*100), 3)}%")
        print("********************************************************************")
```
